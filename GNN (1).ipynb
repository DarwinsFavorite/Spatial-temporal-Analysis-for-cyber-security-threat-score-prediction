```python
import numpy as np
import pandas as pd

df['date'] = pd.to_datetime(df['date'])

# Step 1: Compute original min & max for each (Source-Destination, Attack Type)
original_min_max = df.groupby(['Source-Destination', 'Attack Type'])['Anomaly Scores'].agg(['min', 'max']).to_dict()

# Step 2: Apply exponential decay
decay_factor = 0.97 # Adjust if needed
time_series_dict = {}

for (pair, thread), group in df.groupby(['Source-Destination', 'Attack Type']):
    ts = group.sort_values(by='date', ascending=False)['Anomaly Scores'].values
    ts_decayed = ts * (decay_factor ** np.arange(len(ts)))  # Apply exponential decay
    time_series_dict[(pair, thread)] = ts_decayed

# Step 3: Map decayed values correctly
ff = df.copy()
ff = ff.sort_values(by=['Source-Destination', 'Attack Type', 'date']).reset_index(drop=True)

# Create group-wise index to correctly map decayed values
ff['group_index'] = ff.groupby(['Source-Destination', 'Attack Type']).cumcount()

# Assign decayed values based on correct index
ff['Anomaly Scores'] = ff.apply(
    lambda row: time_series_dict.get((row['Source-Destination'], row['Attack Type']), [np.nan] * len(ff))[row['group_index']]
    if row['group_index'] < len(time_series_dict.get((row['Source-Destination'], row['Attack Type']), []))
    else np.nan,
    axis=1
)

# Step 4: Rescale anomaly scores using original min-max for each (pair, thread type)
def rescale_anomaly_scores(group):
    pair, thread = group['Source-Destination'].iloc[0], group['Attack Type'].iloc[0]
    min_val, max_val = original_min_max['min'][(pair, thread)], original_min_max['max'][(pair, thread)]

    if max_val == min_val:  # Avoid division by zero
        group['Anomaly Scores'] = min_val
    else:
        group['Anomaly Scores'] = (group['Anomaly Scores'] - group['Anomaly Scores'].min()) / (group['Anomaly Scores'].max() - group['Anomaly Scores'].min() + 1e-6)
        group['Anomaly Scores'] = group['Anomaly Scores'] * (max_val - min_val) + min_val

    return group

ff = ff.groupby(['Source-Destination', 'Attack Type'], group_keys=False).apply(rescale_anomaly_scores)

# ✅ Now, `ff["Anomaly Scores"]` contains properly rescaled anomaly scores after decay, without NaN!
print(ff)
```

```python
import pandas as pd
import numpy as np
import networkx as nx
import torch
from torch_geometric.utils import from_networkx
from dtaidistance import dtw
from sklearn.feature_selection import mutual_info_regression
from joblib import Parallel, delayed
import itertools

# ---------------------- LOAD DATA ----------------------
# Ensure 'ff' is pre-loaded with 'Source-Destination', 'Attack Type', 'Anomaly Scores', 'date'
ff['date'] = pd.to_datetime(ff['date'])
ff = ff.sort_values(by=['Source-Destination', 'Attack Type', 'date']).reset_index(drop=True)

# Create dictionary of time series per (Source-Destination, Attack Type)
time_series_dict = {
    (pair, attack): group['Anomaly Scores'].values
    for (pair, attack), group in ff.groupby(['Source-Destination', 'Attack Type'])
}

# Convert time series dict to list for indexing
pairs = list(time_series_dict.keys())
ts_list = list(time_series_dict.values())

# ---------------------- COMPUTE DTW & MI IN PARALLEL ----------------------
import pandas as pd
import numpy as np
import networkx as nx
import torch
from torch_geometric.utils import from_networkx
from dtaidistance import dtw
from sklearn.feature_selection import mutual_info_regression
from joblib import Parallel, delayed
import itertools

# ... (rest of your code)

def compute_dtw_mi(i, j):
    """Compute DTW distance and Mutual Information (MI) between two time series."""
    ts1, ts2 = ts_list[i], ts_list[j]

    # Compute DTW
    dtw_dist = dtw.distance_fast(ts1, ts2)

    # Compute MI safely
    min_len = min(len(ts1), len(ts2))

    # Check if time series have enough data points for MI calculation
    if min_len < 3:  # If too short, assign 0 to MI
        mi_score = 0
    else:
        # Use a smaller number of neighbors if necessary
        n_neighbors = min(3, min_len - 1)  # Ensure n_neighbors < n_samples
        mi_score = mutual_info_regression(
            ts1[:min_len].reshape(-1, 1), ts2[:min_len], n_neighbors=n_neighbors
        )[0]

    return (pairs[i], pairs[j], dtw_dist, mi_score)

# ... (rest of your code)

# Parallel computation
edges = Parallel(n_jobs=-1, backend="loky")(delayed(compute_dtw_mi)(i, j)
            for i, j in itertools.permutations(range(len(ts_list)), 2))

# ---------------------- NORMALIZE EDGE FEATURES ----------------------
max_dtw = max(e[2] for e in edges) + 1e-6
max_mi = max(e[3] for e in edges) + 1e-6

edges = [(src, dst, dtw_dist / max_dtw, mi_score / max_mi) for src, dst, dtw_dist, mi_score in edges]

# ---------------------- BUILD GRAPH ----------------------
G = nx.DiGraph()
G.add_nodes_from(pairs)

# Add node features
for i, pair in enumerate(pairs):
    G.nodes[pair]['features'] = ts_list[i].tolist()

# Add edges with DTW and MI attributes
for src, dst, dtw_norm, mi_norm in edges:
    G.add_edge(src, dst, dtw_distance=dtw_norm, mi_score=mi_norm)

# ---------------------- CONVERT TO PyTorch Geometric Format ----------------------
graph_data = from_networkx(G)

# Pad node features to max length
max_len = max(len(data['features']) for _, data in G.nodes(data=True))
for _, data in G.nodes(data=True):
    data['features'] = np.pad(data['features'], (0, max_len - len(data['features'])), 'constant')

# Convert node features to tensor
graph_data.x = torch.tensor([data['features'] for _, data in G.nodes(data=True)], dtype=torch.float)

# Convert edge attributes to tensor
edge_features = [[data['dtw_distance'], data['mi_score']] for _, _, data in G.edges(data=True)]
graph_data.edge_attr = torch.tensor(edge_features, dtype=torch.float32)

# ---------------------- SAVE GRAPH ----------------------
torch.save(graph_data, "graph_data.pth")
print("Graph saved as 'graph_data.pth' successfully!")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GATConv
from torch_geometric.data import Data
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Define a Graph Attention Network (GAT) model without LSTM
class TimeSeriesGAT(nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels, num_time_steps_to_predict=1,
                 heads=4, dropout=0.2):
        super(TimeSeriesGAT, self).__init__()

        # Graph Attention Layers
        self.gat1 = GATConv(in_channels, hidden_channels, heads=heads, dropout=dropout)
        self.gat2 = GATConv(hidden_channels * heads, hidden_channels * 2, heads=2, dropout=dropout)
        self.gat3 = GATConv(hidden_channels * 2 * 2, hidden_channels, heads=1, dropout=dropout)

        # Attention mechanism for time series feature aggregation
        self.attention = nn.Sequential(
            nn.Linear(hidden_channels, hidden_channels // 2),
            nn.ReLU(),
            nn.Linear(hidden_channels // 2, 1)
        )

        # Final prediction layer
        self.pred_layer = nn.Linear(hidden_channels, out_channels * num_time_steps_to_predict)

        self.num_time_steps = num_time_steps_to_predict
        self.out_channels = out_channels

    def forward(self, x, edge_index, edge_attr=None):
        # Apply first GAT layer
        x = F.relu(self.gat1(x, edge_index, edge_attr))
        x = F.dropout(x, p=0.2, training=self.training)

        # Apply second GAT layer
        x = F.relu(self.gat2(x, edge_index, edge_attr))
        x = F.dropout(x, p=0.2, training=self.training)

        # Apply third GAT layer
        x = F.relu(self.gat3(x, edge_index, edge_attr))

        # Apply attention mechanism over features
        attn_weights = self.attention(x)
        attn_weights = F.softmax(attn_weights, dim=0)
        x = x * attn_weights

        # Final prediction
        x = self.pred_layer(x)

        # Reshape output to [num_nodes, out_channels, num_time_steps]
        x = x.view(-1, self.out_channels, self.num_time_steps)

        return x

# Class to handle data preprocessing with proper scaling
class TimeSeriesDataProcessor:
    def __init__(self, graph_data, pairs):
        self.graph_data = graph_data
        self.pairs = pairs
        self.scalers = {}
        self.original_data = graph_data.x.clone()  # Store original data

    def preprocess(self, prediction_horizon=1):
        """
        Preprocess time series data, including:
        1. Scaling per node time series
        2. Creating input-target pairs
        """
        num_nodes = self.graph_data.x.shape[0]
        time_series_length = self.graph_data.x.shape[1]

        # Initialize scaled data tensor
        scaled_data = torch.zeros_like(self.graph_data.x)

        # Scale each node's time series individually
        for i in range(num_nodes):
            # Get time series for this node
            ts = self.graph_data.x[i].numpy().reshape(-1, 1)

            # Create and fit scaler
            scaler = MinMaxScaler(feature_range=(0, 1))
            scaled_ts = scaler.fit_transform(ts)

            # Store scaled time series
            scaled_data[i] = torch.tensor(scaled_ts.flatten(), dtype=torch.float32)

            # Store scaler for inverse transform later
            self.scalers[i] = scaler

        # Use all time points except the last few as features
        X = scaled_data[:, :-prediction_horizon]

        # Use the last time points as targets
        y = scaled_data[:, -prediction_horizon:]

        # Create a new Data object with scaled data
        processed_data = Data(
            x=X,
            y=y,
            edge_index=self.graph_data.edge_index,
            edge_attr=self.graph_data.edge_attr,
            original_x=self.original_data  # Store original data for reference
        )

        return processed_data

    def inverse_transform_predictions(self, scaled_predictions):
        """
        Transform scaled predictions back to original scale
        """
        num_nodes = len(scaled_predictions)
        predictions = []

        for i in range(num_nodes):
            # Get scaler for this node
            scaler = self.scalers[i]

            # Reshape to 2D array for inverse transform
            scaled_pred = scaled_predictions[i].numpy().reshape(-1, 1)

            # Inverse transform to original scale
            original_pred = scaler.inverse_transform(scaled_pred)

            # Store original scale prediction
            predictions.append(torch.tensor(original_pred.flatten(), dtype=torch.float32))

        return torch.stack(predictions)

    def get_original_series(self, node_idx):
        """
        Get original (unscaled) time series for a node
        """
        return self.original_data[node_idx]

# Function to train the model
def train_model(model, data, epochs=100, lr=0.001):
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    criterion = nn.MSELoss()

    train_losses = []

    model.train()
    for epoch in range(epochs):
        optimizer.zero_grad()

        # Forward pass
        out = model(data.x, data.edge_index, data.edge_attr)
        loss = criterion(out.view(-1), data.y.view(-1))

        # Backward pass
        loss.backward()
        optimizer.step()

        train_losses.append(loss.item())

        if (epoch + 1) % 10 == 0:
            print(f'Epoch {epoch+1}/{epochs}, Loss: {loss.item():.6f}')

    # Plot training loss
    plt.figure(figsize=(10, 5))
    plt.plot(train_losses)
    plt.title('Training Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.savefig('training_loss.png')
    print(f"Training loss plot saved to 'training_loss.png'")

    return model

# Function to evaluate the model
def evaluate_model(model, data, data_processor):
    model.eval()
    with torch.no_grad():
        # Get predictions in normalized scale
        scaled_pred = model(data.x, data.edge_index, data.edge_attr)
        scaled_pred = scaled_pred.view(scaled_pred.shape[0], -1)
        scaled_actual = data.y

        # Calculate metrics on normalized scale
        mse_scaled = mean_squared_error(scaled_actual.numpy(), scaled_pred.numpy())
        mae_scaled = mean_absolute_error(scaled_actual.numpy(), scaled_pred.numpy())

        print(f'Normalized scale - Test MSE: {mse_scaled:.6f}, MAE: {mae_scaled:.6f}')

        # Transform back to original scale
        original_pred = data_processor.inverse_transform_predictions(scaled_pred)

        # Get original scale targets
        original_actual = []
        for i in range(data.y.shape[0]):
            scaler = data_processor.scalers[i]
            original_y = scaler.inverse_transform(data.y[i].numpy().reshape(-1, 1))
            original_actual.append(torch.tensor(original_y.flatten(), dtype=torch.float32))
        original_actual = torch.stack(original_actual)

        # Calculate metrics on original scale
        mse_original = mean_squared_error(original_actual.numpy(), original_pred.numpy())
        mae_original = mean_absolute_error(original_actual.numpy(), original_pred.numpy())

        print(f'Original scale - Test MSE: {mse_original:.6f}, MAE: {mae_original:.6f}')

    return original_pred, original_actual

# Function to perform cross-validation with proper scaling
def cross_validate(model_class, data_processor, graph_data, k=5, epochs=100, lr=0.001):
    """
    Perform k-fold cross-validation on the model with proper scaling
    """
    num_nodes = graph_data.x.shape[0]
    fold_size = num_nodes // k

    mse_scores = []
    mae_scores = []

    for i in range(k):
        print(f"\n{'='*50}")
        print(f"Cross-validation fold {i+1}/{k}")
        print(f"{'='*50}")

        # Split data into train and validation
        val_indices = list(range(i * fold_size, (i + 1) * fold_size))
        train_indices = list(set(range(num_nodes)) - set(val_indices))

        # Process data with scaling
        processed_data = data_processor.preprocess(prediction_horizon=1)

        # Create masks
        train_mask = torch.zeros(num_nodes, dtype=torch.bool)
        train_mask[train_indices] = True

        val_mask = torch.zeros(num_nodes, dtype=torch.bool)
        val_mask[val_indices] = True

        # Initialize model
        in_channels = processed_data.x.shape[1]
        hidden_channels = 64
        out_channels = processed_data.y.shape[1]

        model = model_class(in_channels, hidden_channels, out_channels)

        # Train model on training set
        print(f"Training fold {i+1}/{k}")
        model = train_model(model, processed_data, epochs=epochs, lr=lr)

        # Evaluate on validation set
        model.eval()
        with torch.no_grad():
            out = model(processed_data.x, processed_data.edge_index, processed_data.edge_attr)
            out = out.view(out.shape[0], -1)

            # Calculate metrics for validation nodes only
            val_pred = out[val_mask]
            val_actual = processed_data.y[val_mask]

            # Convert back to original scale
            original_val_pred = []
            original_val_actual = []

            for j, idx in enumerate(val_indices):
                # Get scaler for this node
                scaler = data_processor.scalers[idx]

                # Convert prediction to original scale
                pred = val_pred[j].numpy().reshape(-1, 1)
                original_pred = scaler.inverse_transform(pred)
                original_val_pred.append(original_pred.flatten())

                # Convert actual to original scale
                actual = val_actual[j].numpy().reshape(-1, 1)
                original_actual = scaler.inverse_transform(actual)
                original_val_actual.append(original_actual.flatten())

            # Calculate metrics on original scale
            original_val_pred = np.array(original_val_pred).flatten()
            original_val_actual = np.array(original_val_actual).flatten()

            mse = mean_squared_error(original_val_actual, original_val_pred)
            mae = mean_absolute_error(original_val_actual, original_val_pred)

            mse_scores.append(mse)
            mae_scores.append(mae)

            print(f'Fold {i+1} - MSE: {mse:.6f}, MAE: {mae:.6f}')

    print(f'\nCross-validation results (original scale):')
    print(f'Average MSE: {np.mean(mse_scores):.6f} ± {np.std(mse_scores):.6f}')
    print(f'Average MAE: {np.mean(mae_scores):.6f} ± {np.std(mae_scores):.6f}')

# Function to predict the next value for each time series of each node
def predict_next_values(model, processed_data, data_processor):
    model.eval()

    with torch.no_grad():
        # Use the entire time series as input to predict next values
        scaled_predictions = model(processed_data.x, processed_data.edge_index, processed_data.edge_attr)
        scaled_predictions = scaled_predictions.view(scaled_predictions.shape[0], -1)

        # Transform predictions back to original scale
        original_predictions = data_processor.inverse_transform_predictions(scaled_predictions)

    return original_predictions

# Function to visualize predictions for each node
def visualize_predictions(data_processor, predictions, pairs, num_to_plot=5):
    plt.figure(figsize=(15, 10))

    # Plot at most num_to_plot nodes
    for i in range(min(num_to_plot, len(pairs))):
        plt.subplot(num_to_plot, 1, i+1)

        # Get original time series
        time_series = data_processor.get_original_series(i).numpy()

        # Create time index
        time_index = np.arange(len(time_series))

        # Plot original time series
        plt.plot(time_index, time_series, 'b-', label='Historical Data')

        # Plot predicted next value(s)
        next_point = len(time_series)
        next_value = predictions[i].item()

        # Plot predicted point with same marker style
        plt.plot(next_point, next_value, 'ro', markersize=8, label='Predicted')

        # Connect the last point of the time series to the prediction with a dashed line
        plt.plot([time_index[-1], next_point], [time_series[-1], next_value], 'r--', alpha=0.7)

        # Adjust y-axis limits to accommodate the prediction
        y_min = min(min(time_series), next_value) * 0.95
        y_max = max(max(time_series), next_value) * 1.05
        plt.ylim(y_min, y_max)

        plt.title(f"Node: {pairs[i]}")
        plt.xlabel('Time Step')
        plt.ylabel('Anomaly Score')
        plt.legend()
        plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig('prediction_visualization.png')
    print(f"Prediction visualization saved to 'prediction_visualization.png'")

# Function to analyze time series statistics and prediction alignment
def analyze_time_series_stats(data_processor, predictions, pairs):
    """
    Analyze statistics of time series and how well predictions align with them
    """
    print("\nTime Series Statistics and Prediction Analysis:")
    print("===============================================")

    # Create DataFrame to store statistics
    stats = []

    for i, pair in enumerate(pairs):
        # Get original time series
        time_series = data_processor.get_original_series(i).numpy()

        # Get prediction
        prediction = predictions[i].item()

        # Calculate statistics
        ts_min = np.min(time_series)
        ts_max = np.max(time_series)
        ts_mean = np.mean(time_series)
        ts_std = np.std(time_series)
        ts_last = time_series[-1]

        # Calculate z-score of prediction relative to time series
        if ts_std > 0:
            pred_zscore = (prediction - ts_mean) / ts_std
        else:
            pred_zscore = 0

        # Check if prediction is within historical range
        in_range = ts_min <= prediction <= ts_max

        # Calculate percentage change from last value
        if ts_last != 0:
            pct_change = (prediction - ts_last) / ts_last * 100
        else:
            pct_change = 0

        # Store statistics
        stats.append({
            'Node': f"{pair}",
            'Min': ts_min,
            'Max': ts_max,
            'Mean': ts_mean,
            'Std': ts_std,
            'Last Value': ts_last,
            'Prediction': prediction,
            'Z-score': pred_zscore,
            'In Range': in_range,
            'Change %': pct_change
        })

    # Convert to DataFrame
    stats_df = pd.DataFrame(stats)

    # Print statistics
    pd.set_option('display.max_rows', None)
    pd.set_option('display.width', 120)
    print(stats_df)

    # Count nodes with predictions in range
    in_range_count = stats_df['In Range'].sum()
    print(f"\nPredictions within historical range: {in_range_count}/{len(pairs)} ({in_range_count/len(pairs)*100:.1f}%)")

    # Plot prediction vs. last value
    plt.figure(figsize=(10, 6))
    plt.scatter(stats_df['Last Value'], stats_df['Prediction'], alpha=0.7)
    plt.plot([stats_df['Last Value'].min(), stats_df['Last Value'].max()],
             [stats_df['Last Value'].min(), stats_df['Last Value'].max()], 'r--')
    plt.xlabel('Last Value')
    plt.ylabel('Prediction')
    plt.title('Prediction vs. Last Value')
    plt.grid(True, alpha=0.3)
    plt.savefig('prediction_vs_last.png')
    print(f"Prediction vs. last value plot saved to 'prediction_vs_last.png'")

    return stats_df

# Function to validate the scale alignment of predictions
def validate_prediction_scale(stats_df):
    """
    Validate how well predictions align with the scale of the time series
    """
    print("\nValidating Prediction Scale Alignment:")
    print("====================================")

    # Check if predictions are within reasonable bounds
    outliers = stats_df[abs(stats_df['Z-score']) > 2]
    if len(outliers) > 0:
        print(f"Found {len(outliers)} predictions with |Z-score| > 2 (potential scale issues):")
        print(outliers[['Node', 'Min', 'Max', 'Mean', 'Prediction', 'Z-score']])
    else:
        print("All predictions have Z-scores within 2 standard deviations of their time series mean.")

    # Check for large percentage changes
    large_changes = stats_df[abs(stats_df['Change %']) > 50]
    if len(large_changes) > 0:
        print(f"\nFound {len(large_changes)} predictions with |Change %| > 50% from last value:")
        print(large_changes[['Node', 'Last Value', 'Prediction', 'Change %']])
    else:
        print("\nNo predictions show changes of more than 50% from the last value.")

    # Overall alignment assessment
    avg_zscore = stats_df['Z-score'].abs().mean()
    in_range_pct = stats_df['In Range'].mean() * 100

    print(f"\nOverall scale alignment metrics:")
    print(f"- Average absolute Z-score: {avg_zscore:.2f}")
    print(f"- Percentage within historical range: {in_range_pct:.1f}%")

    if avg_zscore < 1 and in_range_pct > 90:
        print("✓ Excellent scale alignment: Predictions follow historical patterns closely.")
    elif avg_zscore < 1.5 and in_range_pct > 80:
        print("✓ Good scale alignment: Most predictions match the scale of historical data.")
    elif avg_zscore < 2 and in_range_pct > 70:
        print("⚠ Fair scale alignment: Predictions generally match scale but with some outliers.")
    else:
        print("⚠ Poor scale alignment: Many predictions may not match the scale of historical data.")

# Main execution
def main():
    # Assume graph_data and pairs are already defined from your code

    # Initialize data processor
    data_processor = TimeSeriesDataProcessor(graph_data, pairs)

    # Preprocess data with scaling
    prediction_horizon = 1
    processed_data = data_processor.preprocess(prediction_horizon)

    # Define model parameters
    in_channels = processed_data.x.shape[1]  # Number of time steps in features
    hidden_channels = 64
    out_channels = prediction_horizon

    # Create the model
    model = TimeSeriesGAT(in_channels, hidden_channels, out_channels, num_time_steps_to_predict=1)

    # Train the model
    print("\nTraining model...")
    model = train_model(model, processed_data, epochs=150, lr=0.005)

    # Evaluate the model
    print("\nEvaluating model...")
    pred, actual = evaluate_model(model, processed_data, data_processor)

    # Cross-validate
    print("\nPerforming cross-validation...")
    cross_validate(TimeSeriesGAT, data_processor, graph_data, k=5, epochs=100, lr=0.005)

    # Predict next values for each node (in original scale)
    predictions = predict_next_values(model, processed_data, data_processor)

    # Print predictions for each node
    print("\nPredictions for each node:")
    for i, pair in enumerate(pairs):
        print(f"Node {pair}: Predicted next value = {predictions[i].item():.4f}")

    # Visualize predictions
    print("\nVisualizing predictions...")
    visualize_predictions(data_processor, predictions, pairs, num_to_plot=5)

    # Analyze time series statistics and prediction alignment
    print("\nAnalyzing time series statistics...")
    stats_df = analyze_time_series_stats(data_processor, predictions, pairs)

    # Validate prediction scale
    validate_prediction_scale(stats_df)

    # Save the model
    torch.save({
        'model_state_dict': model.state_dict(),
        'scalers': data_processor.scalers  # Save scalers for future use
    }, 'time_series_gat_model.pt')
    print("\nModel and scalers saved to 'time_series_gat_model.pt'")

if __name__ == "__main__":
    main()
```

```python
import torch
import torch.nn.functional as F
from torch_geometric.nn import GATConv
import numpy as np

# Define GAT Model
class GATModel(torch.nn.Module):
    def __init__(self, in_channels, hidden_dim, out_channels):
        super(GATModel, self).__init__()
        self.conv1 = GATConv(in_channels, hidden_dim, heads=4, concat=True)
        self.conv2 = GATConv(hidden_dim * 4, out_channels, heads=1, concat=False)

    def forward(self, x, edge_index, edge_attr):
        x = self.conv1(x, edge_index, edge_attr[:, 0])  # Using DTW as edge_attr
        x = F.relu(x)
        x = self.conv2(x, edge_index, edge_attr[:, 1])  # Using Granger causality as edge_attr
        return x  # Output is next predicted anomaly score

# Load graph data (Ensure `graph_data` is created)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Use the entire time series as node features
X_tensor = graph_data.x.to(device)  # All time series as input
Y_tensor = X_tensor[:, -1].unsqueeze(1)  # Predict next anomaly score (last column)

# Define model
in_channels = X_tensor.shape[1]  # Full time series length as input
hidden_channels = 64
out_channels = 3  # Predicting scores for 3 attack types

model = GATModel(in_channels, hidden_channels, out_channels).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)
criterion = torch.nn.MSELoss()

graph_data = graph_data.to(device)

# Training Loop
epochs = 100
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    out = model(X_tensor, graph_data.edge_index, graph_data.edge_attr)  # Full time series as input
    loss = criterion(out, Y_tensor)  # Predict next anomaly score
    loss.backward()
    optimizer.step()

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item()}")

# Make Predictions for Next Time Step
model.eval()
with torch.no_grad():
    predictions = model(X_tensor, graph_data.edge_index, graph_data.edge_attr)

print("Predictions for next time step (Malware, DDoS, Intrusion):")
print(predictions.cpu().numpy())
```

```python
import pandas as pd
import numpy as np
import networkx as nx
# First make sure to install node2vec
# !pip install node2vec
from node2vec import Node2Vec

def load_and_preprocess_data(lstm_file, gnn_file):
    """
    Load and preprocess LSTM and GNN model outputs.
    """
    # Load data using the provided file paths
    lstm_df = pd.read_csv(lstm_file)
    gnn_df = pd.read_csv(gnn_file)

    # Parse GNN 'Pair' column - handle potential format issues
    try:
        gnn_df['Pair'] = gnn_df['Pair'].apply(eval)
    except:
        # If eval fails, the data may already be in the right format or needs different parsing
        print("Warning: Could not parse 'Pair' column using eval. Check the data format.")

    # Extract components with error handling
    try:
        gnn_df['Source_State'] = gnn_df['Pair'].apply(lambda x: x[0].split('-')[0])
        gnn_df['Dest_State'] = gnn_df['Pair'].apply(lambda x: x[0].split('-')[1])
        gnn_df['Attack_Type'] = gnn_df['Pair'].apply(lambda x: x[1])
    except Exception as e:
        print(f"Error parsing Pair column: {e}")
        print("Please check the format of your 'Pair' column. Expected format: [('State1-State2', 'AttackType')]")
        # Print a sample for debugging
        print("Sample of Pair column:", gnn_df['Pair'].head())
        raise

    # Ensure LSTM data is numeric with better column name handling
    try:
        lstm_df['Prediction'] = pd.to_numeric(lstm_df['Prediction'], errors='coerce')
        # Check if the column is 'Severity Score' or 'Severity_Score'
        if 'Severity Score' in lstm_df.columns:
            lstm_df['Severity_Score'] = pd.to_numeric(lstm_df['Severity Score'], errors='coerce')
        elif 'Severity_Score' in lstm_df.columns:
            lstm_df['Severity_Score'] = pd.to_numeric(lstm_df['Severity_Score'], errors='coerce')
        else:
            print("Warning: Severity Score column not found")
            lstm_df['Severity_Score'] = np.nan
    except Exception as e:
        print(f"Error processing LSTM data: {e}")
        raise

    return lstm_df, gnn_df

def generate_graph_embeddings(gnn_df, dimensions=64, walk_length=30, num_walks=200, workers=4):
    """
    Generate node embeddings using Node2Vec.
    """
    G = nx.Graph()

    # Check if the required columns exist
    if not all(col in gnn_df.columns for col in ['Source_State', 'Dest_State', 'Predicted_Anomaly_Score']):
        if 'Predicted_Anomaly_Score' not in gnn_df.columns and 'Anomaly_Score' in gnn_df.columns:
            gnn_df['Predicted_Anomaly_Score'] = gnn_df['Anomaly_Score']
        else:
            raise ValueError("Missing required columns in GNN dataframe. Required: 'Source_State', 'Dest_State', 'Predicted_Anomaly_Score'")

    # Clean anomaly scores - replace infinite or NaN values with a default value
    gnn_df['Clean_Score'] = gnn_df['Predicted_Anomaly_Score'].copy()

    # Replace inf/-inf values with NaN first
    gnn_df['Clean_Score'].replace([np.inf, -np.inf], np.nan, inplace=True)

    # Check for NaN values and replace them
    nan_count = gnn_df['Clean_Score'].isna().sum()
    if nan_count > 0:
        print(f"Warning: Found {nan_count} NaN/infinite values in anomaly scores. Replacing with median.")
        # Replace NaN with median or a default value
        median_score = gnn_df['Clean_Score'].median()
        if pd.isna(median_score):  # If median is also NaN
            gnn_df['Clean_Score'].fillna(1.0, inplace=True)
        else:
            gnn_df['Clean_Score'].fillna(median_score, inplace=True)

    # Add edges to graph with clean weights
    for _, row in gnn_df.iterrows():
        source = row['Source_State']
        dest = row['Dest_State']
        weight = row['Clean_Score']

        # Final safety check
        if not np.isfinite(weight):
            weight = 1.0

        G.add_edge(source, dest, weight=weight)

    # Check if graph is empty
    if len(G.nodes()) == 0:
        raise ValueError("Graph is empty. Check the GNN dataframe for valid data.")

    # Verify no NaN or infinite weights in the graph
    for u, v, d in G.edges(data=True):
        if not np.isfinite(d['weight']):
            print(f"Warning: Edge ({u}, {v}) still has non-finite weight. Setting to 1.0")
            G[u][v]['weight'] = 1.0

    # Generate embeddings
    try:
        node2vec = Node2Vec(G, dimensions=dimensions, walk_length=walk_length, num_walks=num_walks, workers=workers)
        model = node2vec.fit(window=10, min_count=1, batch_words=4)
        embeddings = {node: model.wv[node] for node in G.nodes()}
    except Exception as e:
        print(f"Error during embedding generation: {e}")
        # Fallback: create default embeddings
        print("Creating default embeddings as fallback")
        embeddings = {node: np.ones(dimensions) for node in G.nodes()}

    return embeddings

def merge_predictions(lstm_df, gnn_df, embeddings):
    """
    Merge LSTM and GNN predictions using graph embeddings.
    """
    # Adjust column name for consistency
    if 'Severity Score' in lstm_df.columns and 'Severity_Score' not in lstm_df.columns:
        lstm_df['Severity_Score'] = lstm_df['Severity Score']

    # Check for required columns
    required_lstm_cols = ['State', 'Attack Type', 'Prediction', 'Severity_Score', 'Column']
    required_gnn_cols = ['Source_State', 'Dest_State', 'Attack_Type', 'Predicted_Anomaly_Score', 'Pair']

    if not all(col in lstm_df.columns for col in required_lstm_cols):
        print(f"Warning: Missing columns in LSTM dataframe. Available columns: {lstm_df.columns.tolist()}")

    if not all(col in gnn_df.columns for col in required_gnn_cols):
        print(f"Warning: Missing columns in GNN dataframe. Available columns: {gnn_df.columns.tolist()}")
        # Try to fix if 'Anomaly_Score' is available instead of 'Predicted_Anomaly_Score'
        if 'Anomaly_Score' in gnn_df.columns and 'Predicted_Anomaly_Score' not in gnn_df.columns:
            gnn_df['Predicted_Anomaly_Score'] = gnn_df['Anomaly_Score']

    # Create dictionaries to store LSTM predictions
    lstm_source_dict = {}
    lstm_dest_dict = {}

    for _, row in lstm_df.iterrows():
        try:
            key = (row['State'], row['Attack Type'])
            value = {'Prediction': row['Prediction'], 'Severity_Score': row['Severity_Score']}
            if row['Column'] == 'Source State':
                lstm_source_dict[key] = value
            elif row['Column'] == 'Destination State':
                lstm_dest_dict[key] = value
        except KeyError as e:
            print(f"Error accessing LSTM row data: {e}")
            continue

    merged_data = []

    for _, row in gnn_df.iterrows():
        try:
            source_key = (row['Source_State'], row['Attack_Type'])
            dest_key = (row['Dest_State'], row['Attack_Type'])

            source_lstm = lstm_source_dict.get(source_key, {'Prediction': np.nan, 'Severity_Score': np.nan})
            dest_lstm = lstm_dest_dict.get(dest_key, {'Prediction': np.nan, 'Severity_Score': np.nan})

            lstm_pred = source_lstm['Prediction'] if pd.notna(source_lstm['Prediction']) else dest_lstm['Prediction']
            lstm_conf = source_lstm['Severity_Score'] if pd.notna(source_lstm['Severity_Score']) else dest_lstm['Severity_Score']

            # Get GNN prediction, with fallback options and clean infinite/NaN values
            if 'Clean_Score' in row:
                gnn_pred = row['Clean_Score']
            elif 'Predicted_Anomaly_Score' in row:
                gnn_pred = row['Predicted_Anomaly_Score']
                if not np.isfinite(gnn_pred):
                    gnn_pred = 1.0
            elif 'Anomaly_Score' in row:
                gnn_pred = row['Anomaly_Score']
                if not np.isfinite(gnn_pred):
                    gnn_pred = 1.0
            else:
                gnn_pred = 1.0
                print("Warning: No anomaly score found for row")

            # Use embeddings to adjust weighting dynamically
            source_emb = embeddings.get(row['Source_State'], np.zeros(64))
            dest_emb = embeddings.get(row['Dest_State'], np.zeros(64))

            # Avoid division by zero
            norm_source = np.linalg.norm(source_emb)
            norm_dest = np.linalg.norm(dest_emb)

            if norm_source > 0 and norm_dest > 0:
                emb_similarity = np.dot(source_emb, dest_emb) / (norm_source * norm_dest)
                # Ensure similarity is in valid range
                emb_similarity = max(-1.0, min(1.0, emb_similarity))
            else:
                emb_similarity = 0

            # Calculate weighted score with safety checks
            adjusted_lstm_weight = 0.4 * (1 + emb_similarity)
            adjusted_gnn_weight = 0.6 * (1 - emb_similarity)
            total_weight = adjusted_lstm_weight + adjusted_gnn_weight

            # Ensure weights are valid
            if not np.isfinite(adjusted_lstm_weight) or not np.isfinite(adjusted_gnn_weight):
                adjusted_lstm_weight = 0.5
                adjusted_gnn_weight = 0.5
                total_weight = 1.0

            # Calculate final score
            if pd.notna(lstm_pred) and pd.notna(gnn_pred):
                # Ensure predictions are finite
                lstm_pred = float(lstm_pred) if np.isfinite(lstm_pred) else 0.5
                gnn_pred = float(gnn_pred) if np.isfinite(gnn_pred) else 0.5
                final_score = (lstm_pred * adjusted_lstm_weight + gnn_pred * adjusted_gnn_weight) / total_weight
            elif pd.notna(lstm_pred):
                final_score = float(lstm_pred) if np.isfinite(lstm_pred) else 0.5
            elif pd.notna(gnn_pred):
                final_score = float(gnn_pred) if np.isfinite(gnn_pred) else 0.5
            else:
                final_score = 0.5

            # Final check that the score is finite
            if not np.isfinite(final_score):
                print(f"Warning: Final score is non-finite. Using default value.")
                final_score = 0.5

            merged_data.append({'Pair': row['Pair'], 'Final_Anomaly_Score': final_score})
        except Exception as e:
            print(f"Error processing row: {e}")
            continue

    return pd.DataFrame(merged_data)

def main(lstm_file, gnn_file, output_file='merged_predictions.csv'):
    """
    Main function to process and merge LSTM & GNN predictions.
    """
    print(f"Loading data from {lstm_file} and {gnn_file}...")
    lstm_df, gnn_df = load_and_preprocess_data(lstm_file, gnn_file)

    print("Generating graph embeddings...")
    embeddings = generate_graph_embeddings(gnn_df)

    print("Merging predictions...")
    merged_df = merge_predictions(lstm_df, gnn_df, embeddings)

    print(f"Saving results to {output_file}...")
    merged_df.to_csv(output_file, index=False)
    print(f"Merged predictions saved to {output_file}")
    return merged_df

# Example usage
# First install node2vec if not already installed
# !pip install node2vec
merged_results = main('/content/final_output (1).csv', '/content/anomaly_predictions.csv')
```
